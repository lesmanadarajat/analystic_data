```python
import pandas as pd
from google.colab import files
import plotly.express as px
uploaded = files.upload()
```

```python
df = pd.read_csv("dataset-stroke.csv")
print(df)
```

```python
labels = df['stroke'].replace({0: 'Tidak Stroke',
                                1: 'Stroke'
                             }).value_counts()
print(labels)


labels = df['stroke'].replace({0: 'Tidak Stroke',
                               1: 'Stroke'
                            }).value_counts(sort=True).index
print(labels)

sizes = df['stroke'].value_counts(sort=True)
print(f"ini sizes: {sizes}")


# PIE CHART
fig = px.pie(names = labels, values = sizes, hole=0.3,
             color_discrete_sequence = ['lightblue', 'red'],
             labels = {'names': 'Diagnosis kesehatan'},
             title = 'Distribusi Diagnosis Kesehatan'
             )

fig.update_traces(textposition='inside',
                  textinfo='percent+label+value')

fig.show()
```

```python
# HISTOGRAM

data_stroke = 1
data_stroke = df[df['stroke'] == data_stroke]
print(data_stroke)

# Histogram
fig = px.histogram(data_stroke, x='age', nbins=50)

# ubah layout visualisasi
fig.update_layout(
    title='Distribusi Usia pada Penderita Stroke',
    xaxis_title='Usia',
    yaxis_title='Jumlah Penderita Stroke',
    showlegend=True
)

fig.show()
```

```python
# jumlah pasien stroke berdasarkan jenis kelamin
data_gender = data_stroke[['gender', 'stroke']].groupby('gender').sum().reset_index()
print(data_gender)


fig = px.pie(data_gender, names = 'gender', values = 'stroke', hole=0.3,
             color_discrete_sequence = ['lightblue', 'red'],
             labels = {'names': 'Diagnosis kesehatan'},
             title = 'Perbandingan Jenis Kelamin pada Penderita Stroke'
             )

fig.update_traces(textposition='inside',
                  textinfo='percent+label+value')

fig.show()
```

```python
# violin plot
fig = px.violin(data_stroke,
                x='gender',
                y='age',
                color = 'gender',
                box=True,
                points='all')

fig.update_layout(title="Persebaran Pasien Stroke berdasarkan Gender",
                  xaxis_title="\nGender",
                  yaxis_title="Age")
fig.show()
```
